# Doing More with Graphs in Python

```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

## Basic Examples

```python
x = np.arange(0, 5, 0.1)
y = np.sin(x)
plt.plot(x, y)
```

```python
x = np.arange(0, 5, 0.1)
y = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
x = np.linspace(-3, 3, 20)
y = x**2
plt.plot(x, y, 'b*--')
```

### Basic Matplotlib Commands

```python
plt.plot(x, y, 'ro--', linewidth=2)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('A graph of $y = x^2$')
```

```python
plt.plot(x, y, 'g+--', linewidth=2, markersize=8)
```

### Creating Multiplots on Same Canvas

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'm--')
plt.title('$x$ is a function of $y$')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'r--')
plt.title('$y$ is a function of $x$')
```

### Introduction to the Object Oriented Method

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.9, 0.8])
axes.plot(x, y, 'b*')
```

```python
fig = plt.figure()
axe1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axe1.plot(x, y, 'b-o', lw=2)
axe1.set_title('Larger Graph')

axe2 = fig.add_axes([0.3, 0.4, 0.4, 0.4])
axe2.plot(y, x, 'r*--',lw=2)
axe2.set_title('Smaller Graph')
```
